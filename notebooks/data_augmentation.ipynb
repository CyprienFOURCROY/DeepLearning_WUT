Rotation

```python
import os
from src import ingest_data
import shutil


BASE_DIR = os.path.abspath("../") 
input_root = os.path.join(BASE_DIR, "data/raw/train")
output_root = os.path.join(BASE_DIR, "data/processed/train_augmented_rotated")


os.makedirs(output_root, exist_ok=True)


for class_name in os.listdir(input_root):

    class_input_folder = os.path.join(input_root, class_name)
    class_output_folder = os.path.join(output_root, class_name)

    if not os.path.isdir(class_input_folder):
        continue

    os.makedirs(class_output_folder, exist_ok=True)

    image_files = sorted(
        [f for f in os.listdir(class_input_folder) if f.lower().endswith(".png")]
    )

    

    for i, img_name in enumerate(image_files):
        

        if i % 1000 == 0 and i != 0 :
            print(f"{i} images processed for class {class_name}")

        input_path = os.path.join(class_input_folder, img_name)
        output_path = os.path.join(class_output_folder, img_name)
        output_path = output_path.replace(".png", "_rotated.png")


        args = [input_path, output_path]

        args.append("--rotate")

        # Copy original file
        output_path = output_path.replace("_rotated.png", ".png")
        shutil.copy2(input_path, output_path)
        # Create augmented version

        ingest_data(standalone_mode=False, args=args)
```

Brightness and contrast

```python
import os
from src import ingest_data
import shutil

BASE_DIR = os.path.abspath("../")
input_root = os.path.join(BASE_DIR, "data/processed/train_augmented_rotated")
output_root = os.path.join(BASE_DIR, "data/processed/train_augmented_rotated_brightness_contrast")
os.makedirs(output_root, exist_ok=True)

for class_name in os.listdir(input_root):
    class_input_folder = os.path.join(input_root, class_name)
    class_output_folder = os.path.join(output_root, class_name)
    
    if not os.path.isdir(class_input_folder):
        continue
        
    os.makedirs(class_output_folder, exist_ok=True)
    
    image_files = sorted(
        [f for f in os.listdir(class_input_folder) if f.lower().endswith(".png")]
    )
    
    l = 0
    for i, img_name in enumerate(image_files):
        l += 1
            
        if i % 1000 == 0 and i != 0:
            print(f"{i} images processed for class {class_name}")
            
        input_path = os.path.join(class_input_folder, img_name)
        
        # Copy original file
        orig_output_path = os.path.join(class_output_folder, img_name)
        shutil.copy2(input_path, orig_output_path)
        
        # Create augmented version
        aug_output_path = os.path.join(class_output_folder, img_name.replace(".png", "_brightness_contrast.png"))
        
        args = [input_path, aug_output_path, "--brightness", "--contrast"]
        ingest_data(standalone_mode=False, args=args)
```
