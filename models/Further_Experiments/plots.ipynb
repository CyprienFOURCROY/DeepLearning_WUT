```python
import os
import json
```

```python
list_size_of_train = [1, 2, 5, 10, 15, 20, 30, 40, 50, 60, 70, 80, 90, 100]
i = 0
list_train_loss = []
list_val_loss = []


list_accuracy_train = []
list_accuracy_val = []

for size in list_size_of_train:
    file_path = os.path.join("logs", f"training_log_DenseNet_proportion{size}.json")

    file_json = json.load(open(file_path))
    dict_key = file_json["data"].keys()
    last_key = list(dict_key)[-1]

    train_loss = file_json["data"][last_key]["summary"]["train_loss"]
    val_loss = file_json["data"][last_key]["summary"]["val_loss"]
    print(file_json["data"][last_key]["summary"])

    train_accuracy = file_json["data"][last_key]["summary"]["train_acc"]
    val_accuracy = file_json["data"][last_key]["summary"]["val_acc"]

    list_train_loss.append(train_loss)
    list_val_loss.append(val_loss)

    list_accuracy_train.append(train_accuracy)
    list_accuracy_val.append(val_accuracy)

    i += 1
```

```python
print("list_accuracy_train", list_accuracy_train)
print("list_accuracy_val", list_accuracy_val)
```

```python
print("list_train_loss: ", list_train_loss)
print("list_val_loss: ", list_val_loss)
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array(list_size_of_train)
y1 = np.array(list_train_loss)
y2 = np.array(list_val_loss)
plt.plot(x, y1, label='train_loss', marker='o')
plt.plot(x, y2, label='val_loss', marker='o')
plt.xlabel('size of train set ( % )')
plt.ylabel('loss')
plt.title('Loss vs Size of Train Set')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array(list_size_of_train)
y1 = np.array(list_accuracy_train)
y2 = np.array(list_accuracy_val)
plt.plot(x, y1, label="train_acc", marker="o")
plt.plot(x, y2, label="val_acc", marker="o")
plt.xlabel("size of train set ( % )")
plt.ylabel("accuracy")
plt.title("accuracy vs Size of Train Set")
plt.legend()
plt.grid(True)

plt.show()
```
